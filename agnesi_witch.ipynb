```python
"""agnesi_witch.ipynb"""

# cell 1
import numpy as np
import matplotlib.pyplot as plt

"""define function f(x)"""
def f(x):
    a = 1 / 2  # set variable
    return a * (1 / (1 + x**2)) 

""""define power series approximation"""
def power_series(x, n_terms):
    result = 0
    for n in range(n_terms):
        coefficient = (-1)**n
        numerator = 1
        denominator = 2**n * np.math.factorial(n)
        term = coefficient * numerator / denominator * x**(2 * n)
        result += term
    return result
```

```python
# cell 2

""" range of x values"""
x_values = np.linspace(-1.5, 1.5, 100)

""" Plot power series and f(x) """
plt.figure(figsize=(10, 6))
plt.plot(x_values, f(x_values), label='Exact f(x)', linewidth=2)

""" Plot more terms of the power series """
for n_terms in range(2, 8):
    approx_values = [power_series(x, n_terms) for x in x_values]
    plt.plot(x_values, approx_values, label=f'{n_terms} terms')

""" plot settings"""
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title("Agnesi's Witch and Power Series Approximations")
plt.legend(loc='best')
plt.grid()
plt.show()
```
