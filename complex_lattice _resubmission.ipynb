```python
# cell 1

import numpy as np
import matplotlib.pyplot as plt

"""define the complex function"""


def f(z):
    return z**2 + 1j * z + 1


"""mesh grid of the complex numbers"""
Re, Im = np.meshgrid(np.linspace(-4, 4, 1000), np.linspace(0, 2, 1000))
n = Re + 1j * Im
# real = Re, imaginary = Im
# generates 1000 numbers for both


"""make a mesh grid"""
F = f(n)
plt.scatter(F.real, F.imag, s=1)

"""plot labels and limits"""
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.xlabel("Re(f(z))")
plt.ylabel("Im(f(z))")
plt.title("Scatter Plot of Complex Function")
plt.gca().set_aspect("equal")  # aspect ratio is equal

plt.show()  # print plot
```
