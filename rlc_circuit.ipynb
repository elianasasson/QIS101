```python
"""rlc_circuit.ipynb"""
# cell 1

import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

""" Data"""
R = 0.1  # Ohms
L = 0.01  # Henries
C = 0.01  # Farads

# Initial conditions
I0 = 1  # Initial current
dI0_dt = 0  # Initial derivative of current


#  second-order homogeneous ODE for the RLC circuit
def rlc_circuit(t, y):  # t is time, y is
    I, dI_dt = y
    d2I_dt2 = -R / L * dI_dt - 1 / (L * C) * I  # ' of current with respect to time
    # = -R / L = voltage drop in circuit
    # L * dI_dt - 1 / (L * C) * I = voltage drop across inductor and capacitor
    return [dI_dt, d2I_dt2]


# set time
t_span = (0, 1)  # From 0 to 1 second

solution = solve_ivp(rlc_circuit, t_span, [I0, dI0_dt])  # solve
```

```python
# cell 2 - plotting

# Plot the current over time
plt.plot(t, I)
plt.xlabel("Time (secs)")
plt.ylabel("Current (Amps)")
plt.title("Current and Time in a RLC Circuit")
plt.grid()

# Display the plot
plt.show()
```
